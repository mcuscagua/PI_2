```python
import os
os.chdir('D:/Compartida/Python 3')
os.getcwd()

import matplotlib.pyplot as plt
plt.style.use('seaborn')
%matplotlib inline

import wlkfwdv6 as wk # Paquete creado en Valores-Bancolombia
import Importar as imp
import numpy as np
import pandas as pd
import pyfolio as pf

import warnings
warnings.filterwarnings("ignore")

import alphalensv3 as al

rebalanceos = ['2019-02-28','2019-03-29','2019-04-30','2019-05-30','2019-06-30','2019-07-30','2019-08-30', '2019-09-30', '2019-10-31']
fin = '2019-10-31'
```

# Estrategia Agresiva

```python
#IMPORTAR DATOS

simbolos = ["XLV","XLF","XLI","XLY","XLE","XLK"]

años = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]

basedatos = imp.raw_pipeline(simbolos,años)
basedatos = basedatos[basedatos.index.get_level_values(0)<=fin]

indice=['SPY']

spy = imp.organizarTodo(indice,años)
spy=spy['SPY'].copy()

spy['Rt']=spy["Close"]/spy["Close"].shift(1)-1
```

```python
def primerdia_sem(df):
    # Obtengo el ultimo dia del mes que esta dentro de un df para cada mes del index
    lista=[]
    j=1
    for i in df:
       if j == 1:
           j+=1
           temp=i
           continue
       else:
#            if temp.isocalendar()[1] != i.isocalendar()[1]: # cada semana
           if temp.month != i.month: # cada mes
#            if temp.month != i.month and i.month in [1,4,7,10]: # cada trimestre
#            if temp.month != i.month and i.month in [12,2,4,6,8,10]: # cada trimestre
               lista.append(temp)
               temp=i
           else:
               temp=i
    lista.append(temp)
    return lista
```

```python
# ++++++++++++++++++#
# ------ MAR  ------#
# ++++++++++++++++++#

pmar=[[20],[200]]

def mar(df,parameters):
    """
    df: Un Dataframe
    parameters: Una tupla con los valores de los parametros
    """
    
    jon =df.copy()
    
    grouper = jon.index.get_level_values('symbol')
    def to_rt(serie):
        return serie/serie.shift(1)-1
    
    # Siempre debe estar la siguiente linea
    jon['ret']=jon.groupby(grouper)['Close'].apply(to_rt)
    jon['ret']=jon['ret'].replace(np.nan,0)
    
    def to_mean(serie,w):
        return serie.rolling(w,min_periods = 1).mean()
    
    jon['FMA']=jon.groupby(grouper)['Close'].apply(to_mean,w = parameters[0])
    jon['SMA']=jon.groupby(grouper)['Close'].apply(to_mean,w = parameters[1])
    
    jon['factor'] = jon['FMA']/jon['SMA'] # distancia entre ambas ma

    return jon
```

```python
dicc_funciones = {}
dicc_parametros = {}

dicc_funciones['mar'] = mar
dicc_parametros['mar'] = pmar
```

```python
dicc_backtest = {}
insample = 24
outsample = 1
size = 6
for key,value in dicc_funciones.items():
    print('++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
    print('                                        {}'.format(key))
    print('++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
    sust=wk.research(basedatos,value,primerdia_sem,dicc_parametros[key],
                     insample,outsample,size,'2010')
    dicc_backtest[key] = sust
```

```python
def tiles(x):
    return pd.qcut(x, 3, labels = False) + 1

grouper = dicc_backtest["mar"].dicc_top_fwd_pipeline['(20, 200)'].index.get_level_values(0)
dicc_backtest["mar"].dicc_top_fwd_pipeline['(20, 200)']['Q'] = dicc_backtest["mar"].dicc_top_fwd_pipeline['(20, 200)'].groupby(grouper)["factor"].apply(tiles)
```

```python
dicc_backtest["mar"].dicc_top_fwd_pipeline['(20, 200)'].tail(24)
```

```python
dicc_principales = {}
for key,value in dicc_backtest.items():
    principal_temp = {}
    principal_temp['principal_temp'] = value.ret_strategy
    principal_temp['ret_sym_in_p'] = value.ret_sym_in_p
    principal_temp['iret_sym_in_p'] = value.iret_sym_in_p
    principal_temp['ret_daily'] = value.ret_daily
    principal_temp['iret_daily'] = value.ret_intraday
    principal_temp['df_scores_sum'] = value.df_scores_sum
    principal_temp['df_scores'] = value.df_scores
    principal_temp['df_best_params'] = value.df_best_params
    principal_temp['sym_in_port'] = value.sym_in_port
    principal_temp['fac_in_port'] = value.fac_in_port
    dicc_principales[key] = principal_temp.copy()
```

```python
dicc_bd = {}
dicc_irets = {}
for key,value in dicc_principales.items(): 
    print('++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
    print('                                        {}'.format(key))
    print('++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
    df_q,df_qs_rets,df_qs_rets_d = al.cuantilesv2(value,3)
    dicc_bd[key],dicc_irets[key] = al.plot_tearsheet(df_q,df_qs_rets,df_qs_rets_d,
                                                     value,key,figsize = (15,7))
```

### Calculo ordinario y Evaluación

```python
largos = dicc_bd['mar']
largos['sign'] = np.where(largos['Q']==2,1,0)
largos['lsret'] = np.where(largos['sign']==1,largos['ret']*largos['sign'],np.nan)
def media(x):
    return x.mean()
grouper = largos.index.get_level_values(0)
largos_def = pd.DataFrame(largos.groupby(grouper)['lsret'].apply(media))
largos_def.columns = ['retls']

bd_def = largos_def
bd_def['ls'] = largos_def['retls']
bd_def['date'] = bd_def.index
bd_def.index = pd.to_datetime(bd_def['date'],format = "%Y-%m-%d", utc = None)
del bd_def['date']; del bd_def['retls']

inic = '2012'
pf.create_simple_tear_sheet(bd_def['ls'].loc[bd_def['ls'].index >= inic], 
                            benchmark_rets = spy['Rt'].loc[spy['Rt'].index >= inic])
```

```python
pipe = sust.dicc_top_fwd_pipeline['(20, 200)']
pipe = pipe.loc[np.in1d(pipe.index.get_level_values(0),np.array(pipe.index.get_level_values(0).unique()[-3:]))]

def qqtiles(x,n):
    return pd.qcut(x, n, labels=False,duplicates='drop') + 1

grouper = pipe.index.get_level_values(0)
pipe['qq'] = pipe.groupby(grouper)['factor'].apply(qqtiles,3)

pipe = pipe.loc[np.in1d(pipe.index.get_level_values(0),rebalanceos)]
pipe = pipe.loc[np.in1d(pipe.index.get_level_values(0),np.array(pipe.index.get_level_values(0).unique()[-1]))]

pipe = pipe.loc[pipe['qq']==2]

largosl = list(pipe.index.get_level_values(1))
```
